What are regular expressions?

* Strings with special syntax
* allow us to match patterns in other strings
* applications of regular expressions
  * find all web links in a document
  * parse email eddresses
  * remove/replace unwanted charachters


```python
import re

# re.match(patternString, String), returns the location of the match

re.match("abc", "abcdefg") # Returns where the first string matches identically in the second string

word_regex = "\w+" # looks for just a singular word
re.match(word_regex, "Hello there") # Will match with the first words in our string

re.search("cd", "abcd") # Searches for a match at ANY point of our string
re.match("cd", "abcd") # starts matching at the BEGINING of the string

```

```python
# Ex:

my_string = "Let's write RegEx!  Won't that be fun?  I sure think so.  Can you find 4 sentences?  Or perhaps, all 19 words?"


# Write a pattern to match sentence endings: sentence_endings
sentence_endings = r"[(.?!)]"

# Split my_string on sentence endings and print the result
print(re.split(sentence_endings, my_string))

# Find all capitalized words in my_string and print the result
capitalized_words = r"[A-Z]\w+"
print(re.findall(capitalized_words, my_string))

# Split my_string on spaces and print the result
spaces = r"\s+"
print(re.split(spaces, my_string))

# Find all digits in my_string and print the result
digits = r"\d+"
print(re.findall(digits, my_string))
```

# Intro to tokenization

* Process of turning a string or document into tokens (small chuncks)
* lots of theories and rules
* One step in preparing a text for NLP
* Common tools are nltk library

Why bother?

* easier to map part of speech
* matching common words
* removoing unwanted tokems

Regex groups

* "or" is represented using |
* you can define a group using ()
* Define explicit charachter ranges using []
* [] used to define a new charachter class for exaple [A-Za-z] will now match any of those letters

```python
import re

match_digits_and_words = r'(\d+|cats)' # pattern is digits or the word cats

re.findall(match_digits_and_words, "He has 11 cats")
```
