# Sorting and subsetting

* df.sort_values(column_name, descending = False) Will sort the dataframe based on that column, by default from greatest to lowest or can change with parameter
    * Can also sort by several columns, in this case you would use a list [] in both the ascending and the column name
* df[columnName] will only display the columns you are requesting. Putting it into a list will allow for several columns

Subsetting based on a condition

* df[columns] >= condition # will create a condition to only show the rows that meet that condition, we can have many conditions at once not just one
  
* We can combine conditions using "& or"
  
* df[column].isin([list of items])
  
* df[column].str.contains("String we want it to contain", "Other optional string we want etc")


```python
# Sorts our dataframet by arrest date from the earliest arrest date to the most recent
df.sort_values("ARREST_DATE", ascending = True )


# shows only some of the columns we want
df[["CASE_ID", "CASE_PARTICIPANT_ID"]]


# only shows the cases commited by men, returning a new dataframe only containing those rows

df[df["GENDER"] == "Male"]

# filters data frames that have First Degree murder, but only that exact variation of the spelling
df[df["DISPOSITION_CHARGED_OFFENSE_TITLE"].isin(["FIRST DEGREE MURDER"])]

# Another way to do this if we wanted to be less specific say we wanted any type of murder we could do this instead
df[df["DISPOSITION_CHARGED_OFFENSE_TITLE"].str.contains("MURDER")]
```
