# What is version control?

* Track files in different states
* Simultaneous file development (Many different devs work on same thing without causing issues)
* Combine different versions of file
* revert changes when need be since tracks all changes

# Why is version control important

* without it, its like cooking something and note writing down the steps you took to replicate it in the future
* Allows for replication of data analysis

# Git vs Github

* scalable (Can be used for small projects to large ones)
* Git is not Github, but its common to use them together

# Part Two 

Storing Data with Git

Commits have 3 parts

1) Commit itself. Contains meta data, author, time etc
2) Tree: Tracks the names and locations in the repo
3) Blob: contain snap shots of the change


Looking at commit history

* git log (Will return in chronological order all the commits made in a repo)
*  Press space bar to show more recent commits and press q to return to the terminal closing the log

*  git show startOfHash8 (output displays log entry then the changes between that commit and current version in the repo)


Viewing changes

* git diff -r HEAD (compares a staged file to the last commited version)
* if we want to compare to a previus commited version we add HEAD~1 second most recent commit etc

git show HEAD~3 () not sure?


Difference bettwen two commits

* git diff (Compares changes between two commits)
git diff hash1 hash2 (returns the difference between those two commits)






# Undoing changes before commiting

Removing from the staging area so it is not commited

* get reset HEAD file.name (will unstage that specific file)

* get reset HEAD (will send all files back and not commit any)

Undo changes to an unstaged file

* git checkout -- file.name (Switches the file to the last version in commit useful to reset progress at last check point)

* git checkout . (will delete all files in the repo and makes them the current version in the last commit)


# Restoring and reverting

*git log -3 file.name (show the changes in that file from the last 3 commits)
* git log --since="month day year" --until="month day year" (will show the log of commits since that day) EX: Apr 2 2022
* git checkout hashnumber file.name this will change the file in our repo to the version commited in that hashnumber commit
* git checkout hashnumber (will change all files in repo to that commited repo version)

Finding and removing files that are in the repo but HAVE not been commited at all
* git clean -n(list all untracked files)
* git clean -f(delates those untracked files)

# Branches

* systematically track multiple versions of files AKA parrelel universe
* Multiple users can work together
* git branch (shows the branches of this project)

* git checkout -b name (makes a new branch and moves us to it)

Difference between branches

* git diff branch1 branch2 (shows difference between two branches)

# Working with Branches

* common to work on different components of a project
* branche allow us to keep making progress concurrently
* So we can test different things while still maintaining core components of our code


Switch Branches
* git checkout branchName (switch branches)

Merging Branches

Why do we merge them? 
    Each branch should be used for a task and then incorporate it to main branch

* git merge branch1 branch2 (will merge the two branches)

# Handling Conflict

What is a conflict?
    When a file in different branches have different content preventing the merge
    Need to resolve the conflict before we can merge

Typically occurs when we change the same file in different branches
Large conflicts?

* prevention is better than cure!
* Use each branch for a specific task
* avoid edditing a file in multiple branches



# Creating repos

Why make a repo

* track versions
* easier collab with others
* store everything

Creating a new repo

* git init repoName (Creates our repo, creates it in the directory we were in when we called the command)
* git status (to confirm it is a git repo)

Converting existing project into a repo

* git innit (makes a new repo. Turns the directory you are in into a repo)
* git status (System automatically detects some files have had other versions and prompts you to commit them?)
  Avoid making a git repo inside a git repo, makes them complicated


Working with remotes 

So far we have worked with local repos, create for local projects

Remote repo: repo stored in the cloud such as github

Benefits: Backed up and can be accessed everywhere
Cloning to local system

* git clone path_to_project newName(Clones a local git repo)
* git clone [url] (clones a repo of github link, system will remember where it came from through a tag)
* git remote -v (will give us more info on where the clone was obtained from)
* git remote add NAME URL (creates remote, useful to define name for easier merge?)

# Gathering from a remote repo

* git fetch repoNAME Localbranch (??)
* git fetch repoNAME otherBranch (Now have those contents)
* git merge repoNAME localBranch (Updates the local file based on the remote file)
* git diff REPONAME branch (will give the difference between the remote file and the local file in that branch)

Pulling from a remote

* remote is often ahead of local repos
* fetch and merge is a common workflow

* git pull repoNAME branch (performs both the fetch and merge)


# Pushing to a remote

* Sends our changes to the remote depositary
* git push remote localBranch (pushes content into remote from local)

Try to pull before pushing since there might have been updates.
