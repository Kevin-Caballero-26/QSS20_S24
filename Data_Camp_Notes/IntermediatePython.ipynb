# Importing, scatterplot and line plot

```python
import matplotlib.pyplot as plt

year = [1950, 1970, 1990, 2010]

pop = [2.519, 3.692, 5.263, 6.972]

# plt.pot(x_axis, y-axis)
plt.plot(year, pop)

# Displays the plot
plt.show()

# plt.scatter(X-axis, Y-axis)
plt.scatter(year, pop)

plt.xscale("linear") # Valid inputs are linear, log, symlog, asinh, logit, function, functionlog 

# Displays the plot
plt.show()
```

# Data Visualization customization

* Many options such as different plot types but also how we customize our plots
* choice depends on data and the story you want to tell

Adding labels, color, title

```python
year = []

for x in range(1950, 2101):
    year.append(x)

pop = []

startingPop = 2.538
for x in range(len(year)):
    pop.append(startingPop)
    startingPop = (startingPop)*1.01


plt.plot(year, pop)

plt.xlabel("Year")
plt.ylabel("Population")
plt.title("World Population")
plt.grid(True) # will add grids to our data


plt.yticks([0, 2, 4, 6, 8, 10, 12], ["0", "2B", "4B", "6B", "8B", "10B", "12B"]) # changes the y scale in parameter 1, then changes label

plt.show()
```

# Dictionaries

* Easy to map things and be able to refrence things quickly without loosing info
* Dictionaries have a key and a value
* Keys must be uniqeu, using the same key will just result in that keys value being updated
* keys should be immutable object such as integers and strings but never a list for example

```python
dict1 = {
    "USA": 900,
    "El Salvador": 400
}


dict1["France"] = 500 # This is how we add something to a dictionary

# to remove a key and its value. We still index with square brackets
del(dict1["France"])

"France" in dict1
dict1
```
